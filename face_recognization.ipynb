```python
import cv2
import face_recognition as fr
import pandas as pd
import numpy as np
```

```python
fd=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
vid = cv2.VideoCapture(0)

try:
    face_db = pd.read_csv("new_dataframe.csv",index_col=0)
    data  ={"NAME":face_db["name"].values.tolist(),
            "ENCODING":face_db["enc"].values.tolist()}
    
except Exception as e:
    print(e)
    data  = {"NAME":[] , "ENCODING":[]}
    
    
    
    
    
    
name = data["NAME"]
enc = data["ENCODING"]



captured=True
counter=0

while captured:   
    flag,img =vid.read()
    if flag:
        
        faces=fd.detectMultiScale(img,1.1,5)
        for x,y,w,h in faces:
            face=img[y-50:y+h+50,x-50:x+w+50].copy()
            # face=img[y:y+h,x:x+w]
            # cv2.rectangle(img,pt1=(x,y),pt2=(x+w,x+h),color=(255,0,0),thickness=3)
            
            if len(faces)==1:
                
                try:
                    
                    fresh_face_enc=fr.face_encodings(face)[0]
                    for ind,fe in enumerate(enc):
                        
                        matched=fr.compare_faces([fresh_face_enc],np.array(eval(fe)))[0]
                        if matched:
                            print("matched")
                            print(name[ind])
                            cv2.putText(img,name[ind],(50,50),cv2.FONT_HERSHEY_COMPLEX,1.5,(0,0,255),4)
                            cv2.rectangle(img,pt1=(x,y),pt2=(x+w,y+h),color=(0,0,255),thickness=2)
                        else:
                            print("not matched")
                except IndexError:
                    print("no face detected in image")
      

        cv2.imshow("webcam_image",img)
        key=cv2.waitKey(1)
        if key==ord("q"):
            break
        
        

    else:
        break
cv2.destroyAllWindows()
vid.release()
```
